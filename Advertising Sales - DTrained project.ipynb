# Training process begin

Creating the train_test_split

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
x_train
```

```python
x_test
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
coeff_df=pd.DataFrame(lr.coef_,x.columns,columns=['Coefficient'])
coeff_df
```

```python
lr.intercept_
```

TO make predictions on the test data

```python
y_pred=lr.predict(x_test)
```

```python
comp_df=pd.DataFrame({'Actual Sales':y_test,'Predicted Sales':y_pred})
comp_df
```

```python
comp_df.shape
```

```python
print('Errors:')
print('Mean Squared Error :',mean_squared_error(y_test,y_pred))
print('Mean Absolute Error :',mean_absolute_error(y_test,y_pred))
print('Root Mean Squared Error :',np.sqrt(mean_squared_error(y_test,y_pred)))
```

# To find coefficient of determination - r2 score

```python
from sklearn.metrics import r2_score
print('Coefficient of determination :',r2_score(y_test,y_pred))
```

```python
lr.score(x_test,y_test).round(3)
```

Our coefficient of determination is greater than 0.7, so our model is not bad. Also our model is some
what close to 1. Now let's see the score of the model on the training data

```python
lr.score(x_train,y_train).round(3)
```

We can see the performance of the model on the training data is close to the performance of the test 
data. Therefore our model is not overfitted
